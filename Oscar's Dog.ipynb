```python
import matplotlib.pyplot as plt
import networkx as nx
G=nx.DiGraph()
G.add_node('Sa',pos=(1,2))
G.add_node('A',pos=(0.5,1))
G.add_node('B',pos=(1.5,1))
G.add_node('F1',pos=(0,0))
G.add_node('F1c',pos=(1,0))
G.add_node('F1c ',pos=(2,0))
G.add_edge('Sa','A',weight=0.4)
G.add_edge('Sa','B',weight=0.6)
G.add_edge('A','F1',weight=0.25)
G.add_edge('A','F1c',weight=0.75)
G.add_edge('B','F1c ',weight=1.0)
pos=nx.get_node_attributes(G,'pos')

red_edges = [('Sa','A'),('A','F1')]
blue_edges = []
edge_colours = ['black' if not edge in red_edges+blue_edges else 'red'
                for edge in G.edges()]
black_edges = [edge for edge in G.edges() if edge not in red_edges+blue_edges]

nx.draw_networkx_labels(G, pos)
labels = nx.get_edge_attributes(G,'weight')
nx.draw_networkx_edge_labels(G,pos,edge_labels=labels)
nx.draw_networkx_nodes(G, pos,  node_color='pink')
nx.draw_networkx_edges(G, pos, edgelist=red_edges, edge_color='r',  arrows=True)
nx.draw_networkx_edges(G, pos, edgelist=black_edges,  arrows=True)

ax = plt.gca()
ax.set_axis_off()
plt.savefig('od-a1.png', dpi = 300)
```
